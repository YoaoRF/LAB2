```python
import matplotlib.pyplot as plt
x = [2012, 2022, 2025, 2028]
y = [42, 43, 45, 47]
x2 = [2012, 2022, 2025, 2028]
y2 = [42, 48, 49, 50]
plt.plot(x, y, label="Perú", color="blue", marker="o")
plt.plot(x2, y2, label="Colombia", color="green", marker="o")
plt.xlabel("Año")
plt.ylabel("Población (millones)")
plt.title("P1: Población Perú vs Colombia")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

salario = pd.read_csv("salario-6.csv")
plt.scatter(salario["edad"], salario["estudio"])
plt.title("P2: edad vs estudio")
plt.xlabel("Edad")
plt.ylabel("Estudio")
plt.grid(True)
plt.show()

print("Correlación Pearson:", salario[["edad","estudio"]].corr().iloc[0,1])
```

```python
plt.scatter(salario["estudio"], salario["ingreso"])
plt.title("P3: estudio vs ingreso")
plt.xlabel("Estudio")
plt.ylabel("Ingreso")
plt.grid(True)
plt.show()

print("Correlación:", salario[["estudio","ingreso"]].corr().iloc[0,1])
```

```python
import pandas as pd

personas2 = pd.read_excel("personas2-5.xlsx")
personas2 = personas2.replace(["ausente", "AUSENTE"], pd.NA)

porcentaje = personas2.isna().mean().mul(100).round(2)
porcentaje = porcentaje.astype(str) + "%"

print(porcentaje.to_string(index=True))

```

```python
import pandas as pd
import numpy as np

df5 = pd.DataFrame({
    'x1': [1, 2, 8, 4, 1, 6],
    'x2': [100, 2, 3, 1, 2, 7],
    'x3': [34, np.nan, np.nan, np.nan, 27, 44],
    'x4': [102, 121, 343, np.nan, 121, 125]
})

medianas = df5.median(numeric_only=True)
df5_imputado = df5.fillna(medianas)
print(df5_imputado)
```

```python
personas2 = pd.read_excel("personas2-5.xlsx").replace(["ausente", "AUSENTE"], pd.NA)
personas2 = personas2.fillna(personas2.mean(numeric_only=True))
print(personas2.isna().mean().mul(100).round(2))
```

```python
df7 = pd.DataFrame({
    'x1': [1, 2, 3, 4, 5, 6],
    'x2': [1, 2, np.nan, 1, np.nan, 7],
    'x3': [32, 30, np.nan, np.nan, 27, 44],
    'x4': [102, np.nan, 343, np.nan, 121, np.nan]
})

medias = df7.mean(numeric_only=True)
df7_imputado = df7.fillna(medias)
print(df7_imputado)
```
