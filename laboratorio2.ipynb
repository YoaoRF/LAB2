```python
import matplotlib.pyplot as plt
x = [2012, 2022, 2025, 2028]
y = [42, 43, 45, 47]

x2 = [2012, 2022, 2025, 2028]
y2 = [42, 48, 49, 50]

plt.plot(x, y, marker='o', label="Perú")
plt.plot(x2, y2, marker='s', label="Colombia")

plt.title("Población Perú vs Colombia")
plt.xlabel("Años")
plt.ylabel("Población (millones)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
from sklearn.datasets import load_iris
import pandas as pd

iris_data = load_iris(as_frame=True)
df = iris_data.frame
df["Species"] = df["target"].map(dict(zip(range(3), iris_data.target_names)))

sns.countplot(data=df, x="Species")
plt.title("Frecuencia de Species en Iris")
plt.show()
```

```python
sns.histplot(df["sepal width (cm)"], bins=15, kde=True)
plt.title("Histograma de Sepal Width")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frecuencia")
plt.show()
```
